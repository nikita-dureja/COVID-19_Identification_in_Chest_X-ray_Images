```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import cv2
%matplotlib inline
```

## Illustrate 5 exemplar images under each category

```python
#Reading data from the merged file to access images
df = pd.read_csv('merged_data_final.csv')
print(df.head(5))

#Filtering covid-19 labels
covid = df.label == "covid-19 pneumonia"
print(df[covid].head(5))

#Displaying 5 images of covid-19 category
for i in range(0,500,100): #specifying this range to display images under different dataset
    path = df[covid]['filename'].iloc[i]
    print(path)
    img = cv2.imread(path, cv2.IMREAD_UNCHANGED)
    plt.title("covid-19 pneumonia")
    plt.imshow(img, cmap = 'gray')
    plt.show()
```

```python
#Filtering non-covid-19 pneumonia labels
noncovid = df.label == "non-covid-19 pneumonia"
print(df[noncovid].head(5))

#Displaying 5 images of non-covid-19 category
for i in range(0,1700,400):#specifying this range to display images under different dataset
    path = df[noncovid]['filename'].iloc[i]
    print(path)
    img = cv2.imread(path, cv2.IMREAD_UNCHANGED)
    plt.title("non-covid-19 pneumonia")
    plt.imshow(img, cmap = 'gray')
    plt.show()
```

```python
#Filtering non-infection labels
noninfection = df.label == "non-infection"
print(df[noninfection].head(5))

#Displaying 5 images of non-covid-19 category
for i in range(0,1450,350):#specifying this range to display images under different dataset
    path = df[noninfection]['filename'].iloc[i]
    print(path)
    img = cv2.imread(path, cv2.IMREAD_UNCHANGED)
    plt.title("non-infection")
    plt.imshow(img, cmap = 'gray')
    plt.show()
```

## Number of images in each category

```python
#Displaying number of images under each category

print('No of images in each category:')
print(df['label'].value_counts())
```

```python
#Illustrating count through bar chart

rt = df["label"].value_counts()

plt.bar(rt.index, rt.values, color=['red','blue','green'])

#Title of the bar chart
plt.title("Number of images under each category")

#x-axis label
plt.xlabel("Image-type")

#y-axis label
plt.ylabel("counts")

#Show plot
plt.show()
```
